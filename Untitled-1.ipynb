```python

import pandas as pd
from scipy import stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url1 = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_220903.txt"
url2 = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_220910.txt"
url3 = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_220917.txt"
url4 = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_220924.txt"
url5 = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_221001.txt"
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
df4 = pd.read_csv(url4)
df5 = pd.read_csv(url5)
```

```python
df = pd.concat([df1, df2, df3, df4, df5])
```
